<h1>Manufacturing Analysis </h1>

```python
# @title Group by mnf
MnfCalor=df[["mfr","calories","rating","vitamins"]]
AvgMnfCalor=MnfCalor.groupby("mfr").mean().reset_index()
AvgMnfCalor.round(2)
```

```python
# @title Average calories by Manufacturer
# bar chart
ax = AvgMnfCalor.plot(kind='bar', x='mfr', y='calories', figsize=(6, 3), title='Calories',legend=False)

# Hide the top and right spines
plt.gca().spines[['top', 'right']].set_visible(False)

# Add data labels
for p in ax.patches:
    height = p.get_height()
    # Format the label to remove decimals
    label = f'{height:.0f}'
    # inserting the bar with label
    ax.annotate(label, (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
# @title Average rating by Manufacturer
# bar chart
ax = AvgMnfCalor.plot(kind='bar', x='mfr', y='rating', figsize=(6, 3), title='Rating',legend=False)

# Hide the top and right spines
plt.gca().spines[['top', 'right']].set_visible(False)

# Add data labels
for p in ax.patches:
    height = p.get_height()
    # Format the label to remove decimals
    label = f'{height:.0f}'
    # inserting the bar with label
    ax.annotate(label, (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')

# Adjust the title position
ax.set_title('Rating', pad=20)
plt.show()
```

```python
# @title Average vitamins by Manufacturer
# bar chart
ax = AvgMnfCalor.plot(kind='bar', x='mfr', y='vitamins', figsize=(6, 3), title='vitamins',legend=False)

# Hide the top and right spines
plt.gca().spines[['top', 'right']].set_visible(False)

# Add data labels
for p in ax.patches:
    height = p.get_height()
    # Format the label to remove decimals
    label = f'{height:.0f}'
    # inserting the bar with label
    ax.annotate(label, (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', xytext=(0, 5), textcoords='offset points')

plt.show()
```

<h1>Type Based on Analysis of Average Calories and Ratings</h1>

```python
# @title ##calories per Type
h1=df[['type','calories','rating']]
h2=h1.groupby('type').mean().round(2).reset_index()
h2
```

```python
# @title Donut chart for type per calories and per rating
# Data for both chart
sizes1 = h2['calories']
labels1 = h2['type']
sizes2 = h2['rating']
labels2 = h2['type']
# two plots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))
# first donut chart
ax1.pie(sizes1, labels=labels1, autopct='%1.1f%%', startangle=90)
ax1.add_artist(plt.Circle((0, 0), 0.70, color='white'))
ax1.set_title('Average Calories by Type')
# second donut chart
ax2.pie(sizes2, labels=labels2, autopct='%1.1f%%', startangle=90)
ax2.add_artist(plt.Circle((0, 0), 0.70, color='white'))
ax2.set_title('Average Rate by Type')

plt.show()
```

<H1> How many prod in every shell

```python
gp_cell_prod=df[['name','shelf']]
cell_name=gp_cell_prod.groupby('shelf').count()
cell_name
```

```python
import matplotlib.pyplot as plt

# Plot bar chart
cell_name.plot(kind='bar', figsize=(6, 3),color='lightcoral', legend=False)

# Add labels and title
plt.xlabel('Shelf')
plt.ylabel('Count')
plt.title('Count of Items per Shelf')

# Show the plot
plt.show()
```

```python
# @title serving cups vs calories
calrPerCup=df[['calories','cups','protein']]
gpcalrPerCup=calrPerCup.groupby('cups').mean().round(2)
gpcalrPerCup
```

```python
#  @title average of protein per cup
gpcalrPerCup.plot(kind='barh', y='protein', figsize=(10, 6))

plt.xlabel('Protein')
plt.ylabel('Number of Cups')
plt.title('Horizontal Bar Chart of Average Protein per Serving Cup')

plt.show()
```
